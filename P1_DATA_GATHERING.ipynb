```python
# using selenium to to automate the scraping process by using the some functonalities like hovering and clicking

#import webdriver from selenium
from selenium import webdriver

# import By from webdriver
from selenium.webdriver.common.by import By

#Import ActionChains. Achion Chains performs complex actions like hovering
from selenium.webdriver.common.action_chains import ActionChains

#import extra libraries
from locale import atof, setlocale, LC_NUMERIC
```

```python
url = 'https://www.yarnplaza.com/'

# the url to scrape the info 

driver = webdriver.Chrome()
# open chrome using webdriver
```

```python
driver.get(url)
# open url using chrome webdriver
```

```python
# actionChains will be used for hovering on element
a = ActionChains(driver)
```

```python
#import numpy

import numpy as np

# modify the settings
setlocale(LC_NUMERIC, 'French_Canada.1252')

# empty list variable
composition = []
 
# empty list variable
needle_size = []
    
# empty list variable    
price = []

# empty list variable
product = []


# list variable called submenus with two out of the three brands given. 
# The two brands have a similar structure for their product view pages.
# composition field and needle size field are both on the same row for these two brands
# The third brand has a different structure for its product item page.


submenus = ["DMC","Drops"]



# list variable called products with the corresponding product names belonging to the two brands above respectively.
products = ["DMC Natura XL","Drops Safran"]




    
# for loop picks the two brands in the submenus variable one after the other. so it begin from index 0 item and ends with on the index 1 item
for sub_menus in range(0,2):
    
    # hover over the menu called Wool & Yarn
    m = driver.find_element(By.LINK_TEXT, "Wool & Yarn")
    a.move_to_element(m).perform()
    
    # find the appropriate element using the passed brand name in the submenus variable
    n = driver.find_element(By.LINK_TEXT, submenus[sub_menus])
    
    # click on the brand name link
    a.move_to_element(n).click().perform()
    
   # find the appropriate element using the passed product name in the products variable 
    
    o = driver.find_element(By.LINK_TEXT, products[sub_menus])
    
     # click on the product link
    a.move_to_element(o).perform()
    a.move_to_element(o).click().perform()
    
    # using xpath get the composition information from the table
    composition_value = driver.find_element(By.XPATH, '//div[@id= "pdetailTableSpecs"] //tbody/tr[4]').text
    
    # put it in the  list variable called composition
    composition.append(composition_value)
    
    # using xpath get the needle size information from the table
    needle_size_value = driver.find_element(By.XPATH, '//div[@id= "pdetailTableSpecs"] //tbody/tr[5]').text
    
    # put it in the  list variable called needle size
    needle_size.append(needle_size_value) 
    
    # using xpath get the price information from the table
    price_value = driver.find_element(By.XPATH, '//div[@id= "ContentPlaceHolder1_upPricePanel"]//child::span[@class="product-price-amount"]').text
    
     # put it in the  list variable called price
    price.append(price_value)
    
    # using xpath get the product information from the table
    product_name = driver.find_element(By.XPATH, '//div[@class= "variants-title"]//child::span[@class="variants-title-txt"]').text
    
     # put it in the  list variable called product
    product.append(product_name)
    
    # end of for loop
  
 # scrape information on the 3rd brand name Stylecraft with product Special DK
    # for this product the composition information is on the 3rd row of the table 
    # unlike the other two products whose composition is on the fourth row 
    # the same scenario for the needle size information

 # hover over the menu called Wool & Yarn
m = driver.find_element(By.LINK_TEXT, "Wool & Yarn")

a.move_to_element(m).perform()

 # find the appropriate element using the brand name called Stylecraft
n = driver.find_element(By.LINK_TEXT, "Stylecraft")

 # click on the brand name link
a.move_to_element(n).click().perform()

 # find the appropriate element using the product called Stylecraft Special DK
o = driver.find_element(By.LINK_TEXT, "Stylecraft Special DK")
a.move_to_element(o).perform()

# click on the product link
a.move_to_element(o).click().perform()

# using xpath get the composition information from the table
composition_value = driver.find_element(By.XPATH, '//div[@id= "pdetailTableSpecs"] // tbody/tr[3]').text

 # put it in the list variable called composition
composition.append(composition_value)

 # using xpath get the needle size information from the table
needle_size_value = driver.find_element(By.XPATH, '//div[@id= "pdetailTableSpecs"] // tbody/tr[4]').text

 # put it in the list variable called needle size
needle_size.append(needle_size_value) 

 # using xpath get the price information from the table
price_value = driver.find_element(By.XPATH, '//div[@id= "ContentPlaceHolder1_upPricePanel"]//child::span[@class="product-price-amount"]').text

     # put it in the empty list variable called price
price.append(price_value)

 # using xpath get the product information from the table
product_name = driver.find_element(By.XPATH, '//div[@class= "variants-title"]//child::span[@class="variants-title-txt"]').text

 # put it in the  list variable called product
product.append(product_name)


# import pandas
# put scraped information into dataframe

import pandas as pd
# import pandas library
billets = pd.DataFrame({
    "Brand and Name" : product,
   "Composition Information": composition,
    "Needle Information" : needle_size,
    "Price":price
    
})

# convert comma being used in the currency to decimal point. eg 3,93 to 3.93

billets['Price_round'] = billets['Price'].apply(lambda v: atof(v.split()[-1]))

# round prices to the nearest whole numbers.

billets["Nearest_Whole_Number_Price"] = billets['Price_round'].apply(lambda x: round(x, 0))


# select the needed columns to display
selection = billets[['Brand and Name', 'Composition Information', 'Needle Information','Nearest_Whole_Number_Price']]

# output
selection



# In case an error such as cannot find element, hover over the Wool and Yarn menu item and under Brands kindly click on Drops
# on the Drops page increase the display to 96. the default display is 24.
# run only this cell again
   
        
        



    
```
